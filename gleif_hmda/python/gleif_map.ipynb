```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

#set display options for pandas dataframes
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
local_path = ""
```

```python
#get lei map to panel from
#load NIC ID walk
nic_ids = pd.read_csv("/Users/roellk/Desktop/HMDA/hmda-panel/2017/output/panel_lei_map.txt", sep="|", dtype="object")
nic_ids.fillna("-1", inplace=True)
print(len(nic_ids))
```

```python
#load GLEIF LEI list
gleif_df = pd.read_csv("../output/lei_info.txt", sep="|", dtype="object")
fill_cols = ["legal_city", "legal_name", "hq_city", "hq_country"]
gleif_df.fillna("-1", inplace=True)
gleif_df = gleif_df[gleif_df.legal_country=="US"]
print(len(gleif_df))
```

```python
nic_ids["gleif_name_tup"] = nic_ids.apply(lambda x: process.extractOne(x.nm_lgl, gleif_df.legal_name), axis=1)
nic_ids.to_csv("../output/name_tup.txt", index=False, sep="|")
```

```python
nic_ids["gleif_city_tup"] = nic_ids.apply(lambda x: process.extractOne(x.city, gleif_df.legal_city), axis=1)
nic_ids.to_csv("../output/city_tup.txt", index=False, sep="|")
```

```python
nic_ids["gleif_address_tup"] = nic_ids.apply(lambda x: process.extractOne(x.street_line1, gleif_df.legal_street1), axis=1)
```

```python
nic_ids.to_csv("../output/address_tup.txt", index=False, sep="|")
```

```python
nic_ids["gleif_name"] = nic_ids.apply(lambda x: x.gleif_name_tup[0], axis=1)
nic_ids["gleif_name_score"] = nic_ids.apply(lambda x: x.gleif_name_tup[1], axis=1)
```

```python
nic_ids["gleif_name_index"] = nic_ids.apply(lambda x: x.gleif_name_tup[2], axis=1)
nic_ids.to_csv("../output/gleif_test.txt", sep="|", index=False)
```
